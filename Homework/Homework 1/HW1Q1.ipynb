```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#Read the image in colour and save it as img using imread function from cv2
img=cv2.imread("/for_watson.png", cv2.IMREAD_COLOR)
print(type(img))
```

```python
plt.imshow(img)
```

```python
# Split the image in 3 channels to examine the image in a more granular way
# CV2 reads image channels in blue, green, red instead of the usual RBG
# Keep in mind when reading and splitting channels
b, g, r = cv2.split(img)
```

```python
# Show all 3 channels one by one to examine the visible patterns
# Use grayscale since the default for matplotlib is viridis
# Threshold the values between 0 and 255

plt.imshow(b, cmap='gray', vmin=0, vmax=255)
print(b)
```

```python
plt.imshow(g, cmap='gray', vmin=0, vmax=255)
print(g)
```

```python
plt.imshow(r, cmap='gray', vmin=0, vmax=255)
print(r)
```

```python
# Full message could be seen by combining red and blue using the viridis colour pallette
# In grayscale it an be seen using only the red channel and thresholding it as seen in the previous code block
# PS: This could be solved with only thresholding and segmenting the image in foreground and background

new_img = r
new_img[320:] = b[320:]
plt.imshow(new_img, cmap='gray', vmin=0, vmax=255)
```
